Step 3: Data Cleaning & Preprocessing

```python
# Step 3: Data Preprocessing

import pandas as pd
from sklearn.preprocessing import LabelEncoder

# --- Load dataset (skip if already loaded) ---
df = pd.read_csv("small_dataset.csv")

# --- Drop unnecessary columns ---
drop_cols = [
    'Unnamed: 0', 'cc_num', 'first', 'last', 'street', 'city', 'state',
    'trans_num', 'unix_time', 'trans_date_trans_time'
]
df.drop(columns=drop_cols, inplace=True, errors='ignore')

# --- Handle missing values (if any) ---
df.fillna(method='ffill', inplace=True)   # Forward fill small missing data
df.fillna(method='bfill', inplace=True)   # Backward fill as backup

# --- Convert date of birth to AGE (numeric) ---
df['dob'] = pd.to_datetime(df['dob'], errors='coerce')
df['age'] = 2020 - df['dob'].dt.year
df.drop(columns=['dob'], inplace=True)

# --- Encode categorical columns ---
categorical_cols = ['merchant', 'category', 'gender', 'job']

for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])

# --- Define features (X) and target (y) ---
X = df.drop('is_fraud', axis=1)
y = df['is_fraud']

# --- Confirm final shape ---
print("‚úÖ Preprocessing Done!")
print("üßæ Dataset Shape after cleaning:", df.shape)
print("üß© Feature columns:", list(X.columns))
print("üéØ Target column:", y.name)

# --- Optional: Save preprocessed data ---
df.to_csv("cleaned_fraud_data.csv", index=False)
```
